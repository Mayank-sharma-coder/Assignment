## Qus9

```python
def count_frequency(arr):
    freq = {}

    for num in arr:
        if num in freq:
            freq[num] += 1
        else:
            freq[num] = 1 ###########**********((((((()))))))

    print(freq)

arr = [1, 2, 2, 3, 1, 4, 2]
count_frequency(arr)
```

```python

arr=[10,20,30,20,40,10,50,30]

def find_duplicates(arr):
    seen = []
    duplicates = []

    for num in arr:
        if num in seen:
            if num not in duplicates:
                duplicates.append(num)
        else:
            seen.append(num)
```
