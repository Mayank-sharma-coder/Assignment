## Ques-8

```python
import pandas as pd


df1 = pd.DataFrame({
    'StudentID': [101, 102, 103, 104, 105],
    'Name': ['Rahul', 'Ankit', 'Mohit', 'Priya', 'Neha'],
    'Department': ['CSE', 'ECE', 'IT', 'CSE', 'ME']
})

df2 = pd.DataFrame({
    'StudentID': [101, 102, 103, 104, 105],
    'Marks': [85, 72, 91, 68, 88],
    'Grade': ['A', 'B', 'A', 'C', 'A']
})


df = pd.merge(df1, df2, on='StudentID')

print("Merged DataFrame:")
print(df)


print("\nStudents with Grade A:")
print(df[df['Grade'] == 'A'])
```
